```python
from video import get_frame_objects

filename = "IMG_0120.MOV"

frame_objects = get_frame_objects("raw_video/" + filename)
```

```python
frame_objects
```

```python
for i, frame in enumerate(frame_objects):
    for j, obj in enumerate(frame):
        print(obj[0])
        if obj[0] == "car":
            cv2.imwrite("raw_plates/" + 
                        filename[:-4] + "_" + str(i) + "_ "+ str(j) + 
                        ".jpg",
                        obj[1])
        elif obj[0] == "person":
            cv2.imwrite("raw_faces/" + 
                        filename[:-4] + "_" + str(i) + "_ "+ str(j) + 
                        ".jpg",
                        obj[1])
```
